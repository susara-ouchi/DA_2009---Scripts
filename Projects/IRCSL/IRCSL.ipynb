```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_ircsl_insurers():
    url = "https://ircsl.gov.lk/registered-insurers/"
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        
        soup = BeautifulSoup(response.content, 'html.parser')
        
        insurers = []
        table = soup.find('table')  # Adjust based on actual website structure
        
        if table:
            rows = table.find_all('tr')
            for row in rows[1:]:  # Skip header row
                cols = row.find_all('td')
                if cols:
                    insurer_data = {
                        'Name': cols[0].text.strip() if len(cols) > 0 else '',
                        'Type': cols[1].text.strip() if len(cols) > 1 else '',
                        'Registration': cols[2].text.strip() if len(cols) > 2 else ''
                    }
                    insurers.append(insurer_data)
        
        df = pd.DataFrame(insurers)
        df.to_csv('ircsl_insurers.csv', index=False)
        print("Scraped data saved to ircsl_insurers.csv")
        return insurers
    
    except requests.RequestException as e:
        print(f"Error fetching the page: {e}")
        return []
    except Exception as e:
        print(f"Error parsing the page: {e}")
        return []

insurers_list = scrape_ircsl_insurers()
for insurer in insurers_list:
    print(insurer)
```
